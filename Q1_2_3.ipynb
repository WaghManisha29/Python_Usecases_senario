```python
def format_phone(phone):
    # Convert to string and remove non-digit characters manually
    phone = str(phone)
    digits = ''.join([ch for ch in phone if ch.isdigit()])

    # Format only if exactly 10 digits
    if len(digits) == 10:
        return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"          # (XXX) XXX-XXXX
        # return f"{digits[:3]}.{digits[3:6]}.{digits[6:]}"          # XXX.XXX.XXXX
        # return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"          # XXX-XXX-XXXX
        # return f"{digits[:3]} {digits[3:6]} {digits[6:]}"          # XXX XXX XXXX
    else:
        return "000-000-0000"


# Apply to 'phone' column
customer_dataframe["phone"]=customer_dataframe["phone"].apply(format_phone)



customer_dataframe.loc[297]
```
